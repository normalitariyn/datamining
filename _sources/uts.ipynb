## **1. KNN**

Mengimport dataset iris dari sklearn

```python
#create a dataframe with all training data except the target column
X = dataset.drop(columns=["Classification"])#check that the target variable has been removed
X.head()
```

```python
# separate target values
y = dataset["Classification"].values  # view target values
y[0:5]
```

```python
# percent_amount_of_test_data = / HUNDRED_PERCENT
percent_amount_of_test_data = 0.2
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = percent_amount_of_test_data, random_state=1, stratify=y)
```

Menentukan nilai k (tetangga terdekat)

```python
amount_of_neighbor = 10
```

```python
# memasukan klasifikasi ke data 
knn = KNeighborsClassifier(n_neighbors = amount_of_neighbor)

# membuat data latih
knn.fit(X_train,y_train)
```

```python
#menampilkan 5 model pertama prediksi pada data uji
knn.predict(X_test)
```

```python
#cek akurasi pemodelan data
knn.score(X_test, y_test)
```

## **2. Desicion Tree**

import data

```python
import pandas as pd
from sklearn import tree
from matplotlib import pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics
```

import dataset

```python
url = 'https://raw.githubusercontent.com/normalitariyn/dataset/main/dataR2%20(1).csv'
data = pd.read_csv(url)
```

menampilkan dataframe 

```python
data.head(-1)
```

```python
# x sebagai variable independent
x = data.drop(['Classification'], axis = 1)
x.head
```

```python
# y sebagai variable dependent
y = data['Classification']
y.head
```

Memisahkan data latih dengan data uji

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, shuffle=True, stratify=y, random_state=0)
```

Proses prediksi dengan algoritma decision tree

```python
# membuat decision tree dengan menggunakan gini
clf = DecisionTreeClassifier(criterion="gini")
clf = clf.fit(x_train,y_train)

# memprediksi data uji
y_pred = clf.predict(x_test)
```

Menampilkan hasil decision tree

```python
plt.figure(figsize=(15,15))
#membuat pohon dengan plot
a = tree.plot_tree(clf,
                   rounded = True,
                   filled = True,
                   fontsize=8)
#menampilkan plot
plt.show()
```

menampilkan akurasi 

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
