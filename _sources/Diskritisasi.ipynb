1.   **Pengertian**

        Diskritisasi yaitu proses mengubah data numerik menjadi data kategorikal 
2.   **Implementasi**
  *   Diskritisasi dengan Equal Width dan Equal Frequency 
      *   Equal width : dikritisasi yang membagi data numerik menjadi beberapa kelompok dengan lebar kelompok yang kurang lebih sama besar. 
      *   Equal frequency : dikritisasi yang membagi data numerik menjadi beberapa kelompok dengan jumlah anggota per kelompok yang kurang lebih sama besar.
  *   Diskritisasi dengan basis Entropy

       Entropy digunakan untuk menentukan yang manakah node yang akan
menjadi pemecah data latih berikutnya










**3. Implementasi Code**




```python
import numpy as np 
import pandas as pd
url = 'https://raw.githubusercontent.com/normalitariyn/mydataset/main/IRIS.csv'
c=pd.read_csv(url)
```



*   Equal Width

    Untuk menghitung equal width, dibutuhkan library pandas dengan menggunakan method "cut". Buat 3 label untuk mengkategorikan batas tiap ukuran petal dan sepal dengan menggunakan small, medium, large.





*sepal length*

```python
# perhitungan atribut sepal length tanpa label

labels = ["small", "medium", "large"]
amount_of_binning = len(labels)
# melakukan binning dengan atribut lebar petal dalam cm (1.4)
binning_interval = pd.cut(c['sepal_length'],amount_of_binning,True).value_counts()
print(binning_interval)
```

```python
#perhitungan atribut lebar  

labels = ["small", "medium", "large"]
amount_of_binning = len(labels)
#melakukan binning dengan atribut lebar petal dalam cm (1.4)
binning_label = pd.cut(c['sepal_length'],amount_of_binning, True, labels)
sepal_binning = binning_label.value_counts()
print(sepal_binning)
```

```python
# menampilkan dataframe sepal_lenght
pd.concat((c["sepal_length"],binning_label), axis=1)
```



*   Equal Frequency

    Untuk menghitung equal frequency, dibutuhkan library pandas dengan menggunakan method "qcut".



*Sepal Length*

```python
# perhitungan atribut sepal length tanpa label

labels = ["small", "medium", "large"]
amount_of_binning = len(labels)
# melakukan binning dengan atribut sepal length
binning_interval = pd.qcut(c['sepal_length'],amount_of_binning).value_counts()
print(binning_interval)
```

```python
# perhitungan atribut sepal length dengan label

binning_label = pd.qcut(c['sepal_length'],amount_of_binning, labels)
l_sepal_binning = binning_label.value_counts()
print(l_sepal_binning)
```

```python
# menampilkan dataframe sepal length
pd.concat((c["sepal_length"],binning_label), axis=1)
```
