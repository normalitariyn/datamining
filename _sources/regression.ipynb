# **Linear Regression**



*Rumus Persamaan Linear*

**y = ax + b**

ket :

y = variabel terikat 

x = variabel bebas 

a = koefisien/gradien 

b = intercept

```python
import numpy as np
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot
X_test, y_test = make_regression(n_samples=20, n_features=1, noise=0.2)
reg = LinearRegression().fit(X_test, y_test)
reg.score(X_test,y_test)
```

```python
x = X_test.tolist()
y = y_test.tolist()
```

```python
#mencari gradien(a)
reg.coef_
```

```python
#mencari b
reg.intercept_
```

```python
#mencari y dengan memasukan x=3 
reg.predict(np.array([[3]]))
```

```python
pyplot.scatter(X_test,y_test)
Y_plot = reg.coef_*X_test+reg.intercept_
pyplot.plot(X_test, Y_plot, color='r')
pyplot.show()
```
