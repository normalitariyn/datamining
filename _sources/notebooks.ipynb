
*   **MEAN** 

    Mean merupakan salah satu ukuran gejala pusat. Mean dapat dikatakan sebagai wakil kumpulan data. Menentukan mean dapat dilakukan dengan cara menjumlahkan seluruh nilai data, kemudian membaginya dengan banyaknya data.

    **Rumus**
$$
\bar{x}=\frac{\sum x_i}{n}
$$

        Keterangan:
  $\bar{x}$ : nilai rata-rata

  $x_i$ : nilai data ke-i

  n       : banyaknya data



```python
# import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

mean = {'Confirmed_Cases':np.mean(c['Confirmed_Cases']),
		'Suspected_Cases': np.mean(c['Suspected_Cases']),'Hospitalized':np.mean(c['Hospitalized']),
    'Travel_History_Yes':np.mean(c['Travel_History_Yes']),'Travel_History_No':np.mean(c['Travel_History_No'])}

df = pd.DataFrame(mean, index=['mean'])
df
```


*   **MEDIAN**

      Median atau nilai tengah adalah salah satu ukuran pemusatan data. Cara berhitung median adalah pertama menyortir segugus data dari yang terkecil sampai terbesar (atau terbaliknya), lalu nilai pengamatan yang tepat di tengah-tengah bila jumlah datanya ganjil, atau rata-rata kedua pengamatan yang di tengah bila jumlah datanya genap, adalah nilai tengah.

      **Rumus**

      Untuk data ganjil:
$$
M e=X_{\frac{n+1}{2}}
$$

      Untuk data genap:
$$
M e=\frac{X_{\frac{n}{2}}+X_{\left(\frac{n}{2}+1\right)}}{2}
$$

```python
median = {'Confirmed_Cases':np.median(c['Confirmed_Cases']),
		'Suspected_Cases': np.median(c['Suspected_Cases']),'Hospitalized':np.median(c['Hospitalized']),
    'Travel_History_Yes':np.median(c['Travel_History_Yes']),'Travel_History_No':np.median(c['Travel_History_No'])}

df = pd.DataFrame(median, index=['median'])
df
```



*   **MODUS**



*   **PERSENTIL**

    Persentil merupakan pembagian data terurut ke dalam 100 bagian sama banyak.

    **Rumus** :

$$
\mathrm{P}_{\mathrm{i}}=\text { data ke }-\frac{i(n+1)}{100}
$$










```python
percentile25 = {'Confirmed_Cases':np.percentile(c['Confirmed_Cases'],25),
		'Suspected_Cases':np.percentile(c['Suspected_Cases'],25),'Hospitalized':np.percentile(c['Hospitalized'],25),
    'Travel_History_Yes':np.percentile(c['Travel_History_Yes'],25),'Travel_History_No':np.percentile(c['Travel_History_No'],25)
    }

df = pd.DataFrame(percentile25, index=['25%'])
df
```


*   **STANDAR DEVIASI**

    Standar Deviasi atau Simpangan Baku adalah suatu nilai yang digunakan dalam menentukan persebaran data pada suatu sampel, sekaligus untuk melihat seberapa dekat data-data tersebut dengan nilai 'mean'

    **Rumus**


$S=\sqrt{\frac{\sum\left(x_i-\bar{x}\right)^2}{n}}$

$S=$ Standar deviasi

$x_i=$ nilai $\mathrm{x} \mathrm {ke-i}$

$\bar{x}=$ nilai rata-rata data

$n=$ jumlah data


```python
std = {'Confirmed_Cases':st.stdev(c['Suspected_Cases']),
		'Hospitalized':st.stdev(c['Hospitalized']),'Travel_History_Yes':st.stdev(c['Travel_History_Yes']),
    'Travel_History_No':st.stdev(c['Travel_History_No'])}

df = pd.DataFrame(std, index=['std'])
df
```

```python
# import library
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# menentukan dataframe
Confirmed_Cases = c['Confirmed_Cases']
Suspected_Cases = c['Suspected_Cases']
Hospitalized = c["Hospitalized"]
Travel_History_Yes = c["Travel_History_Yes"]
Travel_History_No = c["Travel_History_No"]


# mencari mean, median dan modus

meanCC = Confirmed_Cases.mean()
medianCC = Confirmed_Cases.median()
modusCC = Confirmed_Cases.mode()

meanSC = Suspected_Cases.mean()
medianSC = Suspected_Cases.median()
modusSC = Suspected_Cases.mode()

meanH = Hospitalized.mean()
medianH = Hospitalized.median()
modusH = Hospitalized.mode()

meanTHY = Travel_History_Yes.mean()
medianTHY = Travel_History_Yes.median()
modusTHY = Travel_History_Yes.mode()

meanTHN = Travel_History_No.mean()
medianTHN = Travel_History_No.median()
modusTHN = Travel_History_No.mode()

# menampilkan ukuran nilai tengah
print("Confirmed Case -->",'Mean: ',meanCC,'Median: ',medianCC,'Modus: ',modusCC[0])
print("Susspected Case -->",'Mean: ',meanSC,'Median: ',medianSC,'Modus: ',modusSC[0])
print("Hospitalized -->",'Mean: ',meanH,'Median: ',medianH,'Modus: ',modusH[0])
print("Travel History Yes -->",'Mean: ',meanTHY,'Median: ',medianTHY,'Modus: ',modusTHY[0])
print("Travel History No -->",'Mean: ',meanTHY,'Median: ',medianTHY,'Modus: ',modusTHY[0])

#membuat tabel
from prettytable import PrettyTable
tabel = PrettyTable(["", "Confirmed Case", "Suspected Case", "Hospitalized", "Travel History Yes", "Travel History No"])
tabel.add_row(["Mean", meanCC, meanSC, meanH, meanTHY, meanTHN])
tabel.add_row(["Median", medianCC, medianSC, medianH, medianTHY, medianTHN])
tabel.add_row(["Modus", modusCC, modusSC, modusH, modusTHY, modusTHN])
print(tabel)
```
